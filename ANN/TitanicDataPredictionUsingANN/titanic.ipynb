```python
def ANN_forTitanic(xtrain,ytrain,h1_units=4,h2_units=4,learning_rate=0.015):
    # hidden layer 1
    np.random.seed(101)
    w1=2*np.random.randn(h1_units,xtrain.shape[1])
    b1=np.random.randn(w1)
    # hidden layer 2
    w2=2*np.random.randn(w1,h1_units)
    b2=np.random.randn(w2)
    # output layer 
    w_out=2*np.random.randn(h2_units)
    b_out=np.random.randn(1)
    loss=[]
    vec_y=[]
    for i in range(0,xtrain.shape[0]):
        x_data=xtrain[i]
        h1=Relu(np.dot(w1,x_data)+b1)
        h2=Relu(np.dot(w2,h1)+b2)
        y=sigmoid(np.dot(w_out,h2)+b_out)
        
        ## backword propogation
        
        delta_out=(y-ytrain[i])*sigmoid(y,derivative=True)
        delta_h2=(np.dot(delta_out,w_out))*Relu(h2,derivative=True)
        delta_h1=(np.dot(delta_h2,w2))*Relu(h1,derivative=True)
        
        # gradient decent
        w_out=w_out-learning_rate*delta_out*h2
        b_out=b_out-learning_rate*delta_out
        
        #hidden layer 2
        w2=w2-learning_rate*np.kron(delta_h2,h1).reshape(h2_units,h1_units)
        b2=b2-learning_rate*delta_h2
        
        #hidden layer 2
        w1=w1-learning_rate*np.kron(delta_h1,y).reshape(h2_units,h1_units)
        b1=b1-learning_rate*delta_h1
        
        #cost
        loss.append((1/2)-(y-ytrain[i])**2)
        vec_y.append(y[0])
    batch_loss=[]
    for i in range(0,10):
        loss_avg=0
        for m in range(0,60):
            loss_avg+=vec_y[60*i+m]/60
        batch_loss.append(loss_avg)
        
    plt.figure(figsize=(10,6))
    plt.plot(np.arange(1,len(batch_loss)+1),batch_loss)
    plt.show()
    return w1,b1,w2,b2,w_out,b_out,loss
```
