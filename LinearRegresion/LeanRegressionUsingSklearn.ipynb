```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('data_for_lr.csv')
data.head(5)
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
data.head(2)
```

```python
data.isnull().sum()
data.shape
```

```python
sns.scatterplot(x=data['x'],y=data['y'])
```

```python
# Spliting tha data
#testing data
training_input=np.array(data.x[0:500]).reshape(500,1)
training_output=np.array(data.y[0:500]).reshape(500,1)
testing_input=np.array(data.x[500:699]).reshape(199,1)
testing_output=np.array(data.y[500:699]).reshape(199,1)
print(testing_input.shape)
print(testing_output.shape)
```

```python
lr_model=LinearRegression()
```

```python
lr_model.fit(training_input,training_output)
```

```python
lr_model.coef_ # return the coeficient
```

```python
lr_model.intercept_ 
```

# prediction using sklearn

```python
model_predictions=lr_model.predict(training_input)
model_predictions.shape
```

# Visualization

```python
plt.plot(training_input,model_predictions,"*",color="red",label="Traing data")
plt.plot(training_input,training_output,"*",color="green",label="Original data")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend("Performance Comparision")
plt.show()
```

# calculate cost

```python
cost=mean_squared_error(training_output,model_predictions)
cost
```
