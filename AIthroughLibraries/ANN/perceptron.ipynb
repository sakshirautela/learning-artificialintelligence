# Model and Helper Functions

```python
def sigmoid(x):
    return (1.0)/(1+np.exp(-x))
```

```python
x=np.array([1,2,3,4,5])
sigmoid(x) # this is called function broadcasting in wi=hich whole array can be passed in function and return value get in aarrray form
```

```python
def predict(X,weights):
    # x(size)-n+1 (+1 is a bias term) and matrix W size is (n*1) vector
    z=np.dot(X,weights)
    prediction=sigmoid(z)
    return prediction
```

```python
def loss(X,Y,weights):
    Y_=predict(X,weights)
    cost=np.mean(-Y*np.log(Y_)-(1-Y)*np.log(1-Y_))
    return cost
```

```python
def update(X,Y,weights,learning_rate):
    Y_=predict(X,weights)
    dw=np.dot(X.T,Y_-Y)
    m=X.shape[0]
    weights=(weights-learning_rate*dw)/(float(m))
    return weights
```

```python
def train(X,Y,leaning_rate=0.01,max_Epoch=100):
    ones=np.ones((X.shape[0],1))
    X=np.hstack((ones,X))
    weights=np.zeros(X.shape[1])
    for epoch in range(max_Epoch):
        weights=update(X,Y,weights,learning_rate=leaning_rate)
        if epoch%10==0:
            l=loss(X,Y,weights)
            print("Epoch Number %d los %4.f(epoch,1)")
    return weights
```

```python
weights=train(X,Y)
```
