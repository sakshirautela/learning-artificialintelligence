```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img=cv.imread('image.png')
plt.imshow(img)
```

```python
img=cv.cvtColor(img,cv.COLOR_RGB2BGR)
plt.imshow(img)
```

# working with web cam

```python
capture=cv.VideoCapture(0)
while True:
    _,img=capture.read()
    img=cv.flip(img,1)
    cv.imshow("Frame",img)
    key=cv.waitKey(10)
    if key==27:
        capture.releade()
        break
```
